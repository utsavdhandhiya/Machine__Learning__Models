```python
import pandas as pd

file_name1 = '/content/drive/MyDrive/tmdb_5000_credits.csv'
df1 = pd.read_csv(file_name1)

file_name2 = '/content/drive/MyDrive/tmdb_5000_movies.csv'
df2 = pd.read_csv(file_name2)

movies = df1.merge(df2,on='title')
movies.info()
```

```python
import ast

def convert(obj):
  # If the object is already a list, return it directly
  if isinstance(obj, list):
    return obj
  # Otherwise, attempt to parse it as a Python literal
  else:
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies['genres'] = movies['genres'].apply(convert)
movies['cast'] = movies['cast'].apply(convert)
movies['crew'] = movies['crew'].apply(convert)
```

```python
movies['tags']= movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']].copy()
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```
