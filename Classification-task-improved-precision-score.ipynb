Logistic Regression

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/drive/MyDrive/Social_Network_Ads.csv'
df=pd.read_csv(file_path)

print(df.describe())
sns.pairplot(df, hue='Purchased')
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Splitting the dataset into test set and training set

from sklearn.model_selection import train_test_split

X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)
```

```python
# Feature scalling

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# Build and Train the Logistic Regression Model

from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state=1) # This line of code build the model
classifier.fit(X_train, y_train)                # This line of code trin the model by .fit
```

```python
# Predict on the new values

print(classifier.predict(sc.transform([[60,0]]))) # apne model ne feature transform kari ne build karyu chhe
                                                      # etle jo apne navi value mate direct predict na kari sakay
                                                      # karan ke jena par build karyu te ane jena par predict karie te same hovu koie
                                                      # etel ahiya apne sc.transform no use kari ne value ne transform kari.
                                                      # .predict hamesha 2D arrey mage etle ahiya apne ([[]]) no use karyo.
```

```python
# Predict on a whole test set

y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

cm= confusion_matrix(y_test, y_pred)
print(f"Confusion matrix :\n{cm}")

accuracy_score = accuracy_score(y_test, y_pred)
print(f"Accuracy Score :{accuracy_score}")

ps = precision_score(y_test, y_pred)
print(f"Precision Score :{ps}")

rs = recall_score(y_test, y_pred)
print(f"Recall Score :{rs}")

print(f"F1 Score :{f1_score(y_test, y_pred)}")
```

K-NN

```python
from sklearn.neighbors import KNeighborsClassifier

# All the sptep of data preprocessing is same as above we have to move forward from the Model Building Step.

classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier.fit(X_train, y_train)
```

```python
# Predict on new dataset/ test dataset

y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

cm = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix: \n{cm}")

accuracy_score = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: \n{accuracy_score}")

precision_score = precision_score(y_test, y_pred)
print(f"Precision Score: \n{precision_score}")

recall_score = recall_score(y_test, y_pred)
print(f"Recall Score: \n{recall_score}")

f1_score = f1_score(y_test, y_pred)
print(f"F1 Score: \n{f1_score}")
```
